```python
# alternative_titles = {}
# alternative_keys = ['Also known as', 
#                     'Arabic', 
#                     'Cantonese', 
#                     'Chinese', 
#                     'French', 
#                     'Hangul', 
#                     'Hebrew', 
#                     'Hepburn', 
#                     'Japanese',
#                     'Literally',
#                     'Mandarin',
#                     'McCune-Reischauer',
#                     'Original title',
#                     'Polish',
#                     'Revised Romanization',
#                     'Romanized',
#                     'Russian',
#                     'Simplified',
#                     'Traditional',
#                     'Yiddish']

# for title in alternative_keys:
#     if title:
#         alternative_titles.append(title)

        
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    for key in ['Also known as', 
                    'Arabic', 
                    'Cantonese', 
                    'Chinese', 
                    'French', 
                    'Hangul', 
                    'Hebrew', 
                    'Hepburn', 
                    'Japanese',
                    'Literally',
                    'Mandarin',
                    'McCune-Reischauer',
                    'Original title',
                    'Polish',
                    'Revised Romanization',
                    'Romanized',
                    'Russian',
                    'Simplified',
                    'Traditional',
                    'Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
            
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
            
    return movie

clean_movies = [clean_movie(movie) for movie in wiki_movies]

wiki_movies_df = pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
